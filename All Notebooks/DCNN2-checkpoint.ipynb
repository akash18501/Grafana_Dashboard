```python
from prometheus_api_client import PrometheusConnect
from influxdb import InfluxDBClient
import datetime
import numpy as np
import threading
import matplotlib.pyplot as plt
import pandas as pd
```

```python
prom = PrometheusConnect(url="http://demo.robustperception.io:9090/", disable_ssl=True)
# client = InfluxDBClient(host='localhost', port=8086)
# client.switch_database('metric_predictions_db')
```

```python
data = []
df = None

def create_dataframe(): 
    global df 
    df = pd.DataFrame(data,columns=["time","metric"])
    df["time"] = [x + datetime.timedelta(hours=5,minutes=30) for x in df["time"]]
    df.set_index("time",inplace=True)

def get_metric(): 
#     threading.Timer(15.0,get_metric).start()
    x = prom.custom_query(query="node_load15[24h:1m]")
    
    for y in x[0]["values"]: 
        json_body = [{
            "measurement":"NodeLoad",
            "tags":{
                "instance":x[0]["metric"]["instance"],
                "job":x[0]["metric"]["job"],
            }, 
            "fields":{
                "metric":float(y[1])
            },
            "time":datetime.datetime.utcfromtimestamp(int(y[0]))
        }]
        data.append([datetime.datetime.utcfromtimestamp(int(y[0])),float(y[1])]),
        create_dataframe()
get_metric()
print(df)
```

```python
train = df.iloc[:1200]
test = df.iloc[1200:]
```

```python
scaled_train = train["metric"].tolist()
scaled_test = test["metric"].tolist()
```

### Time Series Generation 

```python
#define generator 
no_input = 240
no_feature = 1

def sequence_generator(sequence, no_input):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end index for this batch
        end_ix = i + no_input
        if end_ix > len(sequence) - 1:
            break;
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X),np.array(y)

X,y = sequence_generator(scaled_train,no_input)
X = X.reshape((X.shape[0],X.shape[1],no_feature))
print(X.shape) 
print(y.shape)
```

```python
#define Model 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
```

```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(no_input, no_feature)))
model.add(Conv1D(filters=64, kernel_size=2, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=2, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=2, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=30, verbose=1)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)), loss_per_epoch)
```

### Testing on trained Data 

```python
last_processed_batch_train = np.array(scaled_train[:no_input])
last_processed_batch_train = last_processed_batch_train.reshape(1,no_input,no_feature)
```

```python
predictions_train = scaled_train[:no_input]

for i in range(len(scaled_train) - no_input): 
    pred_value_train = model.predict(last_processed_batch_train)[0]
    predictions_train.append(pred_value_train[0])
    
    last_processed_batch_train = np.delete(last_processed_batch_train,(0),axis=1)
    last_processed_batch_train = np.append(last_processed_batch_train,[[pred_value_train]],axis=1)
```
